```python
 # Your code here
 # Your code here
import pandas as pd

# Load the dataset
file_path = "/content/Assignment1_denton_housing.csv"
df = pd.read_csv(file_path)

# Sorting data by year
df = df.sort_values(by="year")

# (1) Calculate year-to-year difference in occupied housing units
print("Yearly difference in occupied housing units:")
occupied_units = df[df["title_field"] == "Occupied Housing Units"]
for i in range(len(occupied_units) - 1):
    first_year = occupied_units.iloc[i]["year"]
    next_year = occupied_units.iloc[i + 1]["year"]
    v1 = occupied_units.iloc[i]["value"]
    v2 = occupied_units.iloc[i + 1]["value"]
    difference = abs(v2 - v1)
    print(f"{first_year} - {next_year}: {difference}")

# (2) Calculate the percentage of vacant and occupied housing units for each year

print("Year Vacant (%) Occupied (%)")

years = df["year"].unique()
for year in years:
    total = df[(df["year"] == year) & (df["title_field"] == "Total Housing Units")]["value"].values[0]
    vacant = df[(df["year"] == year) & (df["title_field"] == "Vacant Housing Units")]["value"].values[0]
    occupied = df[(df["year"] == year) & (df["title_field"] == "Occupied Housing Units")]["value"].values[0]

    vacant_pct = round((vacant / total) * 100, 2)
    occupied_pct = round((occupied / total) * 100, 2)
    print(f"{year} {vacant_pct}%  {occupied_pct}%")

# (3) Find the year with the highest vacant and occupied housing units
vacant_units = df[df["title_field"] == "Vacant Housing Units"]
occ_units = df[df["title_field"] == "Occupied Housing Units"]

max_vacant = vacant_units.loc[vacant_units["value"].idxmax()]
max_occupied = occ_units.loc[occ_units["value"].idxmax()]


print(f"Vacant Housing Units: {max_vacant['year']} ({max_vacant['value']})")
print(f"Occupied Housing Units: {max_occupied['year']} ({max_occupied['value']})")

# (4) Calculate the harmonic mean of total housing units
total_units = df[df["title_field"] == "Total Housing Units"]["value"]
harmonic_mean = round(len(total_units) / sum(1 / total_units), 2)

print("Harmonic Mean:", harmonic_mean)
```

```python
# EDGE CASE--> Handling edge cases for CSV files involves addressing a few potential issues that might arise during file reading, writing, or data manipulation.
import csv
import os

class Product:
    def __init__(self, product_id, name, category, quantity, price):
        self.product_id = product_id
        self.name = name
        self.category = category
        self.quantity = quantity
        self.price = price

    def to_dict(self):
        return {
            "product_id": self.product_id,
            "name": self.name,
            "category": self.category,
            "quantity": self.quantity,
            "price": self.price
        }

    def value(self):
        return self.quantity * self.price

class InventoryManager:
    def __init__(self, filename="inventory.csv"):
        self.filename = filename
        self.inventory = {}
        self.load_inventory()

    def load_inventory(self):
        """ Load inventory from the CSV file, handling errors gracefully. """
        if not os.path.exists(self.filename):
            print(f"File '{self.filename}' not found. Creating new inventory.")
            self.populate_initial_inventory()
            self.save_inventory()
            return

        try:
            with open(self.filename, mode='r') as file:
                reader = csv.DictReader(file)
                for row in reader:
                    # Ensure the product data is valid
                    try:
                        product = Product(
                            row["product_id"],
                            row["name"],
                            row["category"],
                            int(row["quantity"]),
                            float(row["price"])
                        )
                        self.inventory[row["product_id"]] = product.to_dict()
                    except ValueError as e:
                        print(f"Skipping invalid product row: {row}. Error: {e}")
        except Exception as e:
            print(f"Error loading inventory: {e}. Creating a new inventory.")
            self.populate_initial_inventory()
            self.save_inventory()

    def save_inventory(self):
        """ Save the inventory to the CSV file, handling errors gracefully. """
        try:
            with open(self.filename, mode='w', newline='') as file:
                fieldnames = ["product_id", "name", "category", "quantity", "price"]
                writer = csv.DictWriter(file, fieldnames=fieldnames)
                writer.writeheader()
                for product in self.inventory.values():
                    writer.writerow(product)
        except Exception as e:
            print(f"Error saving inventory: {e}")

    def add_product(self, product):
        if product.product_id in self.inventory:
            print("Product ID already exists!")
            return
        self.inventory[product.product_id] = product.to_dict()
        self.save_inventory()
        print("Product added successfully!")

    def delete_product(self, product_id):
        if product_id in self.inventory:
            del self.inventory[product_id]
            self.save_inventory()
            print("Product deleted successfully!")
        else:
            print("Product not found!")

    def update_product(self, product_id, **kwargs):
        if product_id in self.inventory:
            for key, value in kwargs.items():
                if key in self.inventory[product_id]:
                    self.inventory[product_id][key] = value
            self.save_inventory()
            print("Product updated successfully!")
        else:
            print("Product not found!")

    def search_product(self, search_term):
        results = [prod for prod in self.inventory.values() if search_term.lower() in prod["name"].lower() or search_term.lower() in prod["category"].lower()]
        return results

    def list_products(self):
        return list(self.inventory.values())

    def restock_product(self, product_id, restock_quantity):
        if product_id in self.inventory:
            self.inventory[product_id]["quantity"] = int(self.inventory[product_id]["quantity"]) + restock_quantity
            self.save_inventory()
            print(f"Product {product_id} restocked successfully!")
        else:
            print("Product not found!")

    def generate_report(self):
        category_totals = {}
        total_value = 0
        for product in self.inventory.values():
            product_value = int(product["quantity"]) * float(product["price"])
            total_value += product_value
            category_totals[product["category"]] = category_totals.get(product["category"], 0) + product_value

        print("\nInventory Report:")
        print(f"Total inventory value: ${total_value:.2f}")
        print("Category Breakdown:")
        for category, value in category_totals.items():
            print(f"{category}: ${value:.2f}")

    def populate_initial_inventory(self):
        sample_products = [
            (f"P{i+1}", f"Product{i+1}", f"Category{(i%5)+1}", (i+1)*2, (i+1)*10) for i in range(50)
        ]
        for p in sample_products:
            self.inventory[p[0]] = {
                "product_id": p[0],
                "name": p[1],
                "category": p[2],
                "quantity": p[3],
                "price": p[4]
            }

if __name__ == "__main__":
    manager = InventoryManager()
    while True:
        print("\nInventory Management System")
        print("1. Add Product")
        print("2. Delete Product")
        print("3. Update Product")
        print("4. Search Product")
        print("5. List Products")
        print("6. Restock Product")
        print("7. Generate Report")
        print("8. Exit")

        choice = input("Enter your choice: ")

        if choice == "1":
            pid = input("Enter Product ID (8 digits): ")
            if len(pid) != 8 or not pid.isdigit():
                print("Invalid Product Id. Product ID must be exactly 8 digits long.")
                continue
            name = input("Enter Product Name: ")
            category = input("Enter Product Category: ")
            quantity = int(input("Enter Quantity: "))
            price = float(input("Enter Price: "))
            manager.add_product(Product(pid, name, category, quantity, price))

        elif choice == "2":
            pid = input("Enter Product ID to delete: ")
            manager.delete_product(pid)

        elif choice == "3":
            pid = input("Enter Product ID to update: ")
            print("Enter new values (press enter to skip any field):")
            name = input("New Name: ") or None
            category = input("New Category: ") or None
            quantity = input("New Quantity: ")
            price = input("New Price: ")
            updates = {}
            if name: updates["name"] = name
            if category: updates["category"] = category
            if quantity: updates["quantity"] = int(quantity)
            if price: updates["price"] = float(price)
            manager.update_product(pid, **updates)

        elif choice == "4":
            term = input("Enter product name or category to search: ")
            results = manager.search_product(term)
            print("Search Results:")
            for r in results:
                print(r)

        elif choice == "5":
            products = manager.list_products()
            print("\nCurrent Inventory:")
            for prod in products:
                print(prod)

        elif choice == "6":
            pid = input("Enter Product ID to restock: ")
            quantity = int(input("Enter restock quantity: "))
            manager.restock_product(pid, quantity)

        elif choice == "7":
            manager.generate_report()

        elif choice == "8":
            print("Exiting system. Goodbye!")
            break
        else:
            print("Invalid choice! Please try again.")
```

```python
import csv
import os

class Product:
    def __init__(self, product_id, name, category, quantity, price):
        self.product_id = product_id
        self.name = name
        self.category = category
        self.quantity = quantity
        self.price = price

    def to_dict(self):
        return {
            "product_id": self.product_id,
            "name": self.name,
            "category": self.category,
            "quantity": self.quantity,
            "price": self.price
        }

class InventoryManager:
    def __init__(self, filename="inventory.csv"):
        self.filename = filename
        self.load_inventory()

    def load_inventory(self):
        """Load inventory from the CSV file."""
        if os.path.exists(self.filename):
            with open(self.filename, mode='r') as file:
                reader = csv.DictReader(file)
                self.inventory = {row['product_id']: row for row in reader}
        else:
            self.inventory = {}

    def save_inventory(self):
        """Save the current inventory to the CSV file."""
        with open(self.filename, mode='w', newline='') as file:
            fieldnames = ['product_id', 'name', 'category', 'quantity', 'price']
            writer = csv.DictWriter(file, fieldnames=fieldnames)
            writer.writeheader()
            for product in self.inventory.values():
                writer.writerow(product)

    def add_product(self, product):
        """Add a product to the inventory if its ID doesn't already exist."""
        if not self.is_valid_product_id(product.product_id):
            print("Invalid Product ID format! Product ID must be 8, 10, or 14 characters long.")
            return
        if product.product_id in self.inventory:
            print(f"Product ID '{product.product_id}' already exists!")
            return
        if product.quantity <= 0:
            print("Quantity must be a positive integer!")
            return
        if product.price <= 0:
            print("Price must be a positive value!")
            return
        self.inventory[product.product_id] = product.to_dict()
        self.save_inventory()
        print(f"Product '{product.name}' added successfully!")

    def delete_product(self, product_id):
        """Delete a product from the inventory."""
        if product_id not in self.inventory:
            print(f"Product ID '{product_id}' not found!")
            return
        del self.inventory[product_id]
        self.save_inventory()
        print(f"Product with ID '{product_id}' deleted successfully!")

    def update_product(self, product_id, **kwargs):
        """Update the details of an existing product."""
        if product_id not in self.inventory:
            print(f"Product ID '{product_id}' does not exist!")
            return

        product = self.inventory[product_id]

        if 'quantity' in kwargs:
            if kwargs['quantity'] <= 0:
                print("Quantity must be a positive integer!")
                return
            product['quantity'] = kwargs['quantity']

        if 'price' in kwargs:
            price = kwargs['price']
            # Handle commas in price input
            price = self.sanitize_price_input(price)
            if price <= 0:
                print("Price must be a positive value!")
                return
            product['price'] = price

        if 'name' in kwargs:
            product['name'] = kwargs['name']

        if 'category' in kwargs:
            product['category'] = kwargs['category']

        self.save_inventory()
        print(f"Product with ID '{product_id}' updated successfully!")

    def is_valid_product_id(self, product_id):
        """Check if the product ID is valid (8, 10, or 14 digits)."""
        return len(product_id) in [8, 10, 14]

    def list_products(self):
        """List all products in the inventory."""
        if not self.inventory:
            print("No products in inventory!")
            return []
        return list(self.inventory.values())

    def generate_report(self):
        """Generate an inventory report showing the total value and category breakdown."""
        if not self.inventory:
            print("No products in inventory!")
            return
        total_value = sum(float(product['price']) * int(product['quantity']) for product in self.inventory.values())
        print(f"Total inventory value: ${total_value:.2f}")

        category_breakdown = {}
        for product in self.inventory.values():
            category = product['category']
            value = float(product['price']) * int(product['quantity'])
            if category in category_breakdown:
                category_breakdown[category] += value
            else:
                category_breakdown[category] = value

        print("\nCategory Breakdown:")
        for category, value in category_breakdown.items():
            print(f"{category}: ${value:.2f}")

    def sanitize_price_input(self, price):
        """Sanitize and remove commas in price input, then convert to float."""
        # Check if price is a string and sanitize
        if isinstance(price, str):
            try:
                sanitized_price = price.replace(',', '')  # Remove commas
                return float(sanitized_price)
            except ValueError:
                print(f"Invalid input for price: {price}. Please enter a valid number.")
                return 0.0
        # If it's already a float, just return it
        elif isinstance(price, (int, float)):
            return float(price)
        else:
            print(f"Invalid input for price: {price}. Please enter a valid number.")
            return 0.0

if __name__ == "__main__":
    manager = InventoryManager()

    while True:
        print("\nInventory Management System")
        print("1. Add Product")
        print("2. Delete Product")
        print("3. Update Product")
        print("4. List Products")
        print("5. Generate Report")
        print("6. Exit")
        choice = input("Enter your choice: ")

        if choice == "1":
            pid = input("Enter Product ID: ")
            name = input("Enter Product Name: ")
            category = input("Enter Product Category: ")
            try:
                quantity = int(input("Enter Quantity: "))
                if quantity <= 0:
                    print("Quantity must be a positive integer!")
                    continue
            except ValueError:
                print("Invalid input! Please enter a number for Quantity.")
                continue

            price = input("Enter Price: ")
            price = manager.sanitize_price_input(price)
            if price <= 0:
                continue

            # Edge case: Empty product name or category
            if not name.strip():
                print("Product name cannot be empty!")
                continue
            if not category.strip():
                print("Product category cannot be empty!")
                continue

            manager.add_product(Product(pid, name, category, quantity, price))

        elif choice == "2":
            pid = input("Enter Product ID to delete: ")
            manager.delete_product(pid)

        elif choice == "3":
            pid = input("Enter Product ID to update: ")
            print("Enter new values (press enter to skip any field):")
            name = input("New Name: ") or None
            category = input("New Category: ") or None
            quantity = input("New Quantity: ")
            price = input("New Price: ")
            updates = {}
            if name: updates["name"] = name
            if category: updates["category"] = category

            # Edge case: Ensure quantity and price are valid numbers
            if quantity:
                try:
                    quantity = int(quantity)
                    if quantity <= 0:
                        print("Quantity must be a positive integer!")
                        continue
                    updates["quantity"] = quantity
                except ValueError:
                    print("Invalid input! Please enter a valid number for Quantity.")
                    continue

            if price:
                price = manager.sanitize_price_input(price)
                if price <= 0:
                    continue
                updates["price"] = price

            manager.update_product(pid, **updates)

        elif choice == "4":
            products = manager.list_products()
            if products:
                print("\nCurrent Inventory:")
                for prod in products:
                    print(prod)

        elif choice == "5":
            manager.generate_report()

        elif choice == "6":
            print("Exiting system. Goodbye!")
            break
        else:
            print("Invalid choice! Please try again.")
```

Above code, tries to handle the folloeing 10 edge cases:
The provided flight booking system handles the following **10 edge cases**:

1. **Empty Origin or Destination**  
   - Ensures that both origin and destination are provided when adding or searching for flights.  
   - Example: If the user leaves the origin or destination blank, an error message is displayed.

2. **Invalid City Name (Non-Alphabetic Characters)**  
   - Prevents users from entering numbers or special characters in city names.  
   - Example: Entering "NewYork123" or "L@ndon" will trigger an error.

3. **Negative or Zero Capacity**  
   - Ensures that flight capacity is always a positive integer.  
   - Example: If the user tries to add a flight with a capacity of `-5` or `0`, an error is displayed.

4. **Duplicate Flight Routes**  
   - Prevents adding flights with the same origin and destination that already exist.  
   - Example: If a flight from **New York → Los Angeles** already exists, another one cannot be added.

5. **Overbooking a Flight**  
   - Ensures that a seat cannot be booked if the flight has reached full capacity.  
   - Example: If a flight has a capacity of 10 and all 10 seats are booked, no further reservations can be made.

6. **Invalid Flight ID**  
   - Ensures that only valid flight IDs are used when reserving, releasing, or canceling seats.  
   - Example: If a user enters a non-existent flight ID, an error message is displayed.

7. **Releasing a Seat When None Are Booked**  
   - Prevents users from releasing a seat if no seats are booked on that flight.
