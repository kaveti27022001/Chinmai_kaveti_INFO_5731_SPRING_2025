```python
#part_3-1
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize
from nltk import pos_tag

def count_words(text):
    words = word_tokenize(str(text))

    # Performing POS tagging
    tagged_words = pos_tag(words)
    n_cnt = 0
    v_cnt = 0
    adj_cnt = 0
    adv_cnt = 0
    for word, tag in tagged_words:
        if tag.startswith('N'):
            n_cnt += 1
        elif tag.startswith('V'):
            v_cnt += 1
        elif tag.startswith('J'):
            adj_cnt += 1
        elif tag.startswith('R'):
            adv_cnt += 1

    # Return the counts
    return n_cnt, v_cnt, adj_cnt, adv_cnt
df = pd.read_csv('/content/Narrators_data_Cleaned.csv')
df[['Nouns', 'Verbs', 'Adjectives', 'Adverbs']] = df['Processed_details'].apply(count_words).apply(pd.Series)
tot_nouns = df['Nouns'].sum()
tot_verbs = df['Verbs'].sum()
tot_adjectives = df['Adjectives'].sum()
tot_adverbs = df['Adverbs'].sum()

print(f"Total Nouns: {tot_nouns}")
print(f"Total Verbs: {tot_verbs}")
print(f"Total Adjectives: {tot_adjectives}")
print(f"Total Adverbs: {tot_adverbs}")

df.to_csv("processed_text_count.csv", index=False)

print(f"CSV file 'processed_text_count' has been created successfully!")
```

```python
#part_3-2
import pandas as pd
import spacy
import benepar
from spacy import displacy
from nltk import Tree
df = pd.read_csv('/content/Narrators_data_Cleaned.csv')

nlp = spacy.load("en_core_web_sm")
benepar.download('benepar_en3')
nlp.add_pipe("benepar", config={"model": "benepar_en3"})


def plot_dependency_tree(doc):
    print("\nDependency Parsing Tree (Text Representation):")
    for token in doc:
        print(f"{token.text} --({token.dep_} → {spacy.explain(token.dep_)})--> {token.head.text}")

    displacy.render(doc, style='dep', jupyter=True, options={'compact': True, 'distance': 90})


def plot_constituency_tree(doc):
    for sent in doc.sents:
        print("\nConstituency Parsing Tree (Text Representation):")
        print(sent._.parse_string)
        tree = Tree.fromstring(sent._.parse_string)
        tree.pretty_print()



processed_sentence = df['Processed_details'][2]
doc = nlp(processed_sentence)

print(f"Sentence: {processed_sentence}")

plot_dependency_tree(doc)
plot_constituency_tree(doc)
```

```python
#part_3-3
import spacy
import pandas as pd
from collections import Counter

nlp = spacy.load("en_core_web_sm")

df = pd.read_csv("/content/Narrators_data_Cleaned.csv")

entity_cnts = {
    "PERSON": 0,
    "ORG": 0,
    "GPE": 0,
    "PRODUCT": 0,
    "DATE": 0
}

def extract_entities(text):
    doc = nlp(text)
    entity = {
        "PERSON": [],
        "ORG": [],
        "GPE": [],
        "PRODUCT": [],
        "DATE": []
    }


    for ent in doc.ents:
        if ent.label_ == "PERSON":
            entity["PERSON"].append(ent.text)
        elif ent.label_ == "ORG":
            entity["ORG"].append(ent.text)
        elif ent.label_ == "GPE":
            entity["GPE"].append(ent.text)
        elif ent.label_ == "PRODUCT":
            entity["PRODUCT"].append(ent.text)
        elif ent.label_ == "DATE":
            entity["DATE"].append(ent.text)


    for label in entity:
        entity_cnts[label] += len(entity[label])

name = []
bio = []
processed_name = []
processed_details = []

for index, row in df.iterrows():
    name_text = row['Name']
    bio_text = row['Bio']
    processed_name_text = row['Processed_name']
    processed_details_text = row['Processed_details']
    name.append(name_text)
    bio.append(bio_text)
    processed_name.append(processed_name_text)
    processed_details.append(processed_details_text)
    extract_entities(processed_name_text)
    extract_entities(processed_details_text)


print("Entity Counts:")
for entity, count in entity_cnts.items():
    print(f"{entity}: {count}")

entity_count_df = pd.DataFrame(list(entity_cnts.items()), columns=['Entity', 'Count'])

entity_count_df.to_csv("entity_counts.csv", index=False)
print("csv is saved")
print(f"\nNames: {name}")
print(f"Processed Names: {processed_name}")
print(f"Bios: {bio}")
print(f"Processed Details: {processed_details}")
```

```python
#Part_2
import pandas as pd
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer

# Download necessary NLTK resources
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt_tab')


# Step 1: Load the CSV file into a DataFrame
df = pd.read_csv('/content/Github_products.csv')

# Step 2: Preprocessing Function for 'product_information'
def preprocess_text(text):
    # Remove special characters, HTML tags, and extra spaces
    text_clean = re.sub(r'<.*?>', '', text)  # Remove HTML tags
    text_clean = re.sub(r'[^a-zA-Z\s]', '', text_clean)  # Remove special characters
    text_clean = text_clean.strip()  # Remove extra whitespace

    # Convert to lowercase
    text_clean = text_clean.lower()

    # Tokenize
    tokens = word_tokenize(text_clean)

    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    tokens = [word for word in tokens if word not in stop_words]

    # Lemmatization
    lemmatizer = WordNetLemmatizer()
    lemmatized_tokens = [lemmatizer.lemmatize(word) for word in tokens]

    return ' '.join(lemmatized_tokens)

# Step 3: Apply preprocessing to the 'product_information' column
df['processed_product_info'] = df['info'].apply(preprocess_text)

# Step 4: Data Quality Operations

# (a) Handle missing values (in both columns)
df['Name'].fillna('Unknown', inplace=True)  # Fill missing product names with 'Unknown'
df['info'].fillna('No information available', inplace=True)  # Fill missing info with placeholder

# (b) Remove duplicates based on product_name and product_information
df.drop_duplicates(subset=['Name', 'info'], keep='first', inplace=True)

# (c) Check for consistency (ensure product names are non-empty, and product info is valid)
# Ensure that no product name is empty
if df['Name'].str.strip().eq('').any():
    print("Warning: There are empty product names.")

# Ensure that product_information is not empty
if df['info'].str.strip().eq('').any():
    print("Warning: There are products with no information.")

# (d) Optional: Ensure any date columns (if any) are in consistent format
# For example, if you have a date column, make sure it's properly formatted
# df['date_column'] = pd.to_datetime(df['date_column'], errors='coerce')

# Step 5: Check cleaned data
print(df.head())

# Optional: Save the cleaned DataFrame to a new CSV file
df.to_csv('cleaned_github_product_data.csv', index=False)

```

```python
#Part_2
import pandas as pd
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer

# Download necessary NLTK resources
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt_tab')


# Step 1: Load the CSV file into a DataFrame
df = pd.read_csv('/content/tweets_deepseekai.csv')

# Step 2: Preprocessing Function for 'tweets'
def preprocess_text(text):
    # Remove special characters, HTML tags, and extra spaces
    text_clean = re.sub(r'<.*?>', '', text)  # Remove HTML tags
    text_clean = re.sub(r'[^a-zA-Z\s]', '', text_clean)  # Remove special characters
    text_clean = text_clean.strip()  # Remove extra whitespace

    # Convert to lowercase
    text_clean = text_clean.lower()

    # Tokenize
    tokens = word_tokenize(text_clean)

    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    tokens = [word for word in tokens if word not in stop_words]

    # Lemmatization
    lemmatizer = WordNetLemmatizer()
    lemmatized_tokens = [lemmatizer.lemmatize(word) for word in tokens]

    return ' '.join(lemmatized_tokens)

# Step 3: Apply preprocessing to the 'tweets' column
df['processed_tweets'] = df['text'].apply(preprocess_text)

# Step 4: Data Quality Operations

# (a) Handle missing values (in both columns)
df['text'].fillna('Unknown', inplace=True)
df['tweet_id'].fillna('Unknown', inplace=True)  # Fill missing tweet_id with 'Unknown'
df['author_id'].fillna('No information available', inplace=True)  # Fill missing author_id with placeholder

# (b) Remove duplicates based on tweet_id, text
df.drop_duplicates(subset=['tweet_id', 'text'], keep='first', inplace=True)

# (c) Check for consistency (ensure text are non-empty and valid)
# Ensure that no text is empty
if df['text'].str.strip().eq('').any():
    print("Warning: There are empty text.")

# Step 5: Check cleaned data
print(df.head())

# Optional: Save the cleaned DataFrame to a new CSV file
df.to_csv('cleaned_tweets_data.csv', index=False)

```
